# 博客來網路書店搜尋python

```python
import requests
import pandas as pd
from lxml import etree
from bs4 import BeautifulSoup

res = requests.get("http://search.books.com.tw/search/query/key/python/cat/all")
print (res.text)
```

```python
#爬取書名
soup = BeautifulSoup(res.text,'lxml')
parse = etree.HTML(res.text)
book = pd.Series(parse.xpath('//img[@class="itemcov"]/@alt'))
book
```

```python
#檢查價格內容
for price in soup.select("span[class='price']"):
    print(price.select('b'))
```

```python
#爬取價格
i = 0
prices = pd.Series()
for price in soup.select("span[class='price']"):
    if(i<book.size):
        if(len(price.select('b'))==1): #只有價格
            prices = prices.append(pd.Series([price.select('b')[0].text])).reset_index(drop=True) 
        elif(len(price.select('b'))==2): #有打折數+價格
            prices = prices.append(pd.Series([price.select('b')[1].text])).reset_index(drop=True) 
        else:
            break
    i+=1
```

```python
#合併成DataFrame
df = pd.DataFrame({'書名':book, '價格': prices})
df[['書名','價格']]
```
